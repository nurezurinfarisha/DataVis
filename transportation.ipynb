```python
import pandas as pd

# Load the dataset
file_path = 'ridership_headline.csv'
ridership_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
ridership_data.head()
```

```python
import matplotlib.pyplot as plt

# Convert the date column to datetime
ridership_data['date'] = pd.to_datetime(ridership_data['date'])

# Set the date as the index for easier plotting
ridership_data.set_index('date', inplace=True)

# Plot the time series data
plt.figure(figsize=(14, 8))

for column in ridership_data.columns:
    plt.plot(ridership_data.index, ridership_data[column], label=column)

plt.title('Daily Ridership Trends')
plt.xlabel('Date')
plt.ylabel('Ridership Numbers')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Resample the data to monthly frequency and calculate the mean for each month
monthly_ridership = ridership_data.resample('M').mean()

# Plot the monthly ridership trends
plt.figure(figsize=(14, 8))

for column in monthly_ridership.columns:
    plt.plot(monthly_ridership.index, monthly_ridership[column], label=column)

plt.title('Monthly Ridership Trends')
plt.xlabel('Date')
plt.ylabel('Average Ridership Numbers')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()

plt.show()
```
