```python
import pandas as pd

# Load the provided CSV files
covid_cases = pd.read_csv('covid_cases.csv')
covid_cases_age = pd.read_csv('covid_cases_age.csv')

# Display the first few rows of each dataframe to understand their structure
covid_cases.head(), covid_cases_age.head()
```

```python
import matplotlib.pyplot as plt

# Merge the two dataframes on 'date' and 'state'
merged_df = pd.merge(covid_cases, covid_cases_age, on=['date', 'state'])

# Convert 'date' column to datetime format
merged_df['date'] = pd.to_datetime(merged_df['date'])

# Group by 'date' to aggregate the data
aggregated_df = merged_df.groupby('date').sum().reset_index()

# Plotting the time series analysis
plt.figure(figsize=(14, 8))

# Total new cases, recovered cases, and active cases
plt.plot(aggregated_df['date'], aggregated_df['cases_new'], label='New Cases', color='blue')
plt.plot(aggregated_df['date'], aggregated_df['cases_recovered'], label='Recovered Cases', color='green')
plt.plot(aggregated_df['date'], aggregated_df['cases_active'], label='Active Cases', color='red')

# Cases in different age groups
plt.plot(aggregated_df['date'], aggregated_df['cases_child'], label='Child Cases', linestyle='--', color='cyan')
plt.plot(aggregated_df['date'], aggregated_df['cases_adolescent'], label='Adolescent Cases', linestyle='--', color='magenta')
plt.plot(aggregated_df['date'], aggregated_df['cases_adult'], label='Adult Cases', linestyle='--', color='yellow')
plt.plot(aggregated_df['date'], aggregated_df['cases_elderly'], label='Elderly Cases', linestyle='--', color='orange')

# Adding titles and labels
plt.title('COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Convert the date column to datetime
covid_cases_age_new['date'] = pd.to_datetime(covid_cases_age_new['date'])

# Plotting trends of cases by age group over time
plt.figure(figsize=(14, 10))

# Define age groups
age_groups = ['cases_child', 'cases_adolescent', 'cases_adult', 'cases_elderly']

for age_group in age_groups:
    plt.plot(covid_cases_age_new['date'], covid_cases_age_new[age_group], marker='o', linestyle='-', label=age_group)

plt.title('Trend of COVID-19 Cases by Age Group Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(title='Age Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
