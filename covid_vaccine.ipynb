```python
import pandas as pd

# Load the datasets
covid_cases_vaxstatus = pd.read_csv('covid_cases_vaxstatus.csv')
covid_cases = pd.read_csv('covid_cases.csv')

# Display the first few rows of each dataframe to understand their structure
covid_cases_vaxstatus_head = covid_cases_vaxstatus.head()
covid_cases_head = covid_cases.head()

covid_cases_vaxstatus_head, covid_cases_head
```

```python
import matplotlib.pyplot as plt

# Visualizing covid_cases dataset
plt.figure(figsize=(12, 6))
plt.plot(covid_cases['date'], covid_cases['cases_new'], label='New Cases')
plt.plot(covid_cases['date'], covid_cases['cases_import'], label='Imported Cases')
plt.plot(covid_cases['date'], covid_cases['cases_recovered'], label='Recovered Cases')
plt.plot(covid_cases['date'], covid_cases['cases_active'], label='Active Cases')
plt.plot(covid_cases['date'], covid_cases['cases_cluster'], label='Cluster Cases')
plt.title('COVID-19 Cases in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualizing covid_cases_vaxstatus dataset
plt.figure(figsize=(12, 6))
plt.stackplot(covid_cases_vaxstatus['date'],
              covid_cases_vaxstatus['cases_unvax'],
              covid_cases_vaxstatus['cases_pvax'],
              covid_cases_vaxstatus['cases_fvax'],
              covid_cases_vaxstatus['cases_boost'],
              labels=['Unvaccinated', 'Partially Vaccinated', 'Fully Vaccinated', 'Boosted'])
plt.title('COVID-19 Cases by Vaccination Status in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Visualizing covid_cases dataset
plt.figure(figsize=(12, 6))
plt.plot(covid_cases['date'], covid_cases['cases_new'], label='New Cases')
plt.plot(covid_cases['date'], covid_cases['cases_import'], label='Imported Cases')
plt.plot(covid_cases['date'], covid_cases['cases_recovered'], label='Recovered Cases')
plt.plot(covid_cases['date'], covid_cases['cases_active'], label='Active Cases')
plt.plot(covid_cases['date'], covid_cases['cases_cluster'], label='Cluster Cases')
plt.title('COVID-19 Cases in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualizing covid_cases_vaxstatus dataset
plt.figure(figsize=(12, 6))
plt.stackplot(covid_cases_vaxstatus['date'],
              covid_cases_vaxstatus['cases_unvax'],
              covid_cases_vaxstatus['cases_pvax'],
              covid_cases_vaxstatus['cases_fvax'],
              covid_cases_vaxstatus['cases_boost'],
              labels=['Unvaccinated', 'Partially Vaccinated', 'Fully Vaccinated', 'Boosted'])
plt.title('COVID-19 Cases by Vaccination Status in Malaysia')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(loc='upper left')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
